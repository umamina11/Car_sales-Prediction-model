```python
import pandas as pd
import numpy as np 

# for data visualization 
import seaborn as sns
import matplotlib.pyplot as plt

#importing the dataset that is gong to be used for anlayis with the .csv extension 
carsales = pd.read_csv('../input/car-sales/Car_sales.csv')
# describing the dataset 
display(carsales.describe())
```

```python
#treating the null values by replaing them 
carsales.replace(np.nan,'0',inplace = True)
# checking for the null values
carsales.isnull().sum()
```

```python
### Value counts of the column - Manufacturer

manufacturer_count = carsales['Manufacturer'].value_counts(dropna = False)
manufacturer_count
```

```python
### Bar graph showing the value counts of the column - Manufacturer

plt.figure(figsize = (20, 6))
sns.barplot(manufacturer_count.index, manufacturer_count.values, alpha = 0.8)
plt.title('Bar graph showing the value counts of the column - Manufacturer')
plt.ylabel('Number of Occurrences', fontsize = 12)
plt.xlabel('Manufacturer', fontsize = 12)
plt.show()
```

```python
mean_price_manufacturer = carsales[['Manufacturer', 'Price_in_thousands']].groupby('Manufacturer', as_index = False).mean()
mean_price_manufacturer
```

```python
# Categorical Variables
```

```python
### Value counts of the column - Vehicle_type

vehicle_count = carsales['Vehicle_type'].value_counts(dropna = False)
vehicle_count
```

```python
### Bar graph showing the value counts of the column - Vehicle_type

sns.barplot(vehicle_count.index, vehicle_count.values, alpha = 0.8)
plt.title('Bar graph showing the value counts of the column - Vehicle type')
plt.ylabel('Number of Occurrences', fontsize = 12)
plt.xlabel('Vehicle type', fontsize = 12)
plt.show()
```

```python
### Mean price per each Vehicle type

mean_price_vehicle = carsales[['Vehicle_type', 'Price_in_thousands']].groupby('Vehicle_type', as_index = False).mean()
mean_price_vehicle
```
